```python
import numpy as np
import matplotlib.pyplot as plt

# Define the unit step function
def unit_step(n):
    return np.where(n >= 0, 1, 0)

# Define delta function (discrete impulse)
def delta(n):
    return np.where(n == 0, 1, 0)

# Function to find positions of delta impulses
def find_delta_positions(signal):
    return np.where(signal != 0)[0]

# Define the time range
n = np.arange(-10, 10)  # Define a range for n

# Define the input signal x[n] using delta functions
x = (delta(n + 1) + 
     2 * delta(n - 1) + 
     3 * delta(n - 1) + 
     4 * delta(n - 1) + 
     5 * delta(n - 1) + 
     6 * delta(n - 1))  # x[n]

# Define the impulse response h[n]
h = (delta(n - 1) + 
     delta(n-1) + 
     delta(n - 1) + 
     2 * delta(n - 1))  # h[n]

# Perform convolution
y = np.convolve(x, h, mode='full')  # Convolution
n_y = np.arange(2 * n[0], 2 * n[-1] + 1)  # Time indices for y

# Find positions and amplitudes of delta impulses in the result
delta_positions = n_y[find_delta_positions(y)]
amplitudes = y[y != 0]

# Plot the convolution result
plt.figure(figsize=(10, 5))
plt.stem(delta_positions, amplitudes, basefmt=" ")  # Removed use_line_collection
plt.title("Representation of Convolution Result y[n]")
plt.xlabel("n (Time Index)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()

# Output results with delta positions
print("y[n] (Result of x[n] * h[n]):", y)
print("Delta positions in y[n]:", delta_positions)
```
